```python
import torch
import numpy as np
from torch import nn
from torch.utils.data import TensorDataset, DataLoader
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch.nn.functional as F

df = pd.read_csv('weatherAUS.csv')

df = df.drop('Date', axis=1)

df = pd.get_dummies(df, drop_first=True)

num_cols = df.iloc[:, :-1].select_dtypes(exclude=['object']).columns
cat_cols = df.iloc[:, :-1].select_dtypes(include=['object']).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode())


x = df.drop('RainTomorrow_Yes', axis=1)
y = df['RainTomorrow_Yes']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2 ,random_state=42)

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)


y_test = y_test.to_numpy(dtype=np.float32)
y_train = y_train.to_numpy(dtype=np.float32)
x_train = torch.from_numpy(x_train).float()
y_train = torch.from_numpy(y_train).float()
x_test = torch.from_numpy(x_test).float()
y_test = torch.from_numpy(y_test).float()


train_data = TensorDataset(x_train, y_train)
train_loader = DataLoader(train_data, batch_size=64)


class NeuralNetwork(nn.Module):
  def __init__(self, input_size):
    super(NeuralNetwork, self).__init__()


    self.fc1 = nn.Linear(input_size, 64)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(64, 32)
    self.relu2 = nn.ReLU()
    self.fc3 = nn.Linear(32, 1)


  def forward(self,x):
    logits = self.fc1(x)
    logits = self.relu(logits)
    logits = self.fc2(logits)
    logits = self.relu2(logits)
    logits = self.fc3(logits)
    return logits


input_size = x_train.shape[1]

model = NeuralNetwork(input_size)

loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)





size = len(train_loader.dataset)

def train(dataloader, model, loss_fn, optimizer):
  model.train()
  for batch, (x , y) in enumerate(dataloader):
      y_pred = model(x)
      y = y.view(-1,1)
      loss = loss_fn(y_pred, y)
      loss.requires_grad = True

      optimizer.zero_grad()
      loss.backward()
      optimizer.step()  # Update weights


      if (batch) % 1000 == 0:
          loss, current = loss.item(), batch * len(x)
          print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")










def test(x_test, y_test, model):
  model.eval()

  with torch.no_grad():

    y_pred = model(x_test)
    y_test = y_test.view(-1,1).float()

    y_pred = (y_pred > 0.5).float()  # Threshold at 0.5 for binary classification

    accuracy = torch.mean(((y_pred == y_test).float()))*100


  print(f"Accuracy: {(accuracy.item()):>0.1f}% \n")






num_epochs = 10

for epoch in range(num_epochs):
  print(f"Epoch {epoch+1}\n-------------------------------")
  train(train_loader, model, loss_fn, optimizer)
  test(x_test, y_test, model)





```
